```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx', sheet_name='raw_data_for_indexes')
columns = ['показатели', 'пк', 'шпд', 'сайт', 
           'облако', 'фин_расчеты', 'экон_задачи', 'доступ_бд', 
           'эл_подписи', 'антивирус', 'спецы_высш', 'спецы_сред'
           ]
df.columns = columns
df
```

```python
df['Index_A'] = round((df['пк']*0.1 + df['шпд']*0.2 + df['сайт']*0.3 + df['облако']*0.4), 1)
df['Index_B'] = round((df['фин_расчеты']-0.3 + df['экон_задачи']*0.4 + df['доступ_бд']*0.3), 1)
df['Index_C'] = round((df['спецы_высш']*0.5 + df['спецы_сред']*0.5), 1)
df['Index_D'] = round((df['эл_подписи']*0.5 + df['антивирус']*0.5), 1)
```

```python
df['INDEX'] = round((df['Index_A']*0.3 + df['Index_B']*0.4 + df['Index_C']*0.2 + df['Index_D']*0.1), 1)
```
